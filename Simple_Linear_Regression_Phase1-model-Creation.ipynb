```python
#Stage 1:  model creation
import pandas as pd # is a python library used for data analysis & data manupulation ,it helps to work with tables,spreadsheets fles & large datasets 
from pandas import read_csv # Used to read data from csv file Stage : Data collection
#print(dir(pd))
```

```python
dataset=pd.read_csv("Salary_Data.csv")# reading csv file
```

```python
dataset # checking the value
```

```python
independent=dataset[["YearsExperience"]] # from dataset fetching only yearsExperince column
```

```python
independent
```

```python
dependent=dataset[["Salary"]]# from dataset fetching only Salary column
dependent
```

```python
#stage: Split train and test
from sklearn.model_selection import train_test_split # sklearn is a pythin library and importing train_test_split
#below line : by assigning input and output values to the method and spliting the trainn and test data
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression # importing LinearRegression class for model creation
regressor=LinearRegression()
#help(regressor.fit)
#stage:   train set
regressor.fit(X_train,Y_train)
#print(dir(regressor))
# stage : model creation
weight=regressor.coef_ #calculating weight
#help(regressor.coef_)
weight
#help(regressor.coef_)
```

```python
bias=regressor.intercept_ #calculating bias
bias
#stage: Test set
y_pred=regressor.predict(X_test) # passing test inputs to get predicted  output
```

```python
#stage: Evaluation metrics
#from sklearn library importing r2_score class to calculate the metrics if value is near to 1 then it is a good model
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred) #passing test output and predicted output from model to calculate r2_score value
```

```python
r_score # checking r_score value
```

```python
import pickle # pickle library used to save the best model
file_name="finalized_model_linear.sav"
```

```python
# stage : Save the best model
pickle.dump(regressor,open(file_name,'wb'))# created model will be saved in file_name 
```

```python
# stage : Deployment
loaded_model=pickle.load(open(file_name,'rb')) # stage :loading the saved model
result=loaded_model.predict([[20]]) # stage: get inputs
```
